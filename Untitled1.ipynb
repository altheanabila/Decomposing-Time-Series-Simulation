```python
import pandas as pd
import statsmodels.api as sm
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomp_df = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', header=0, parse_dates=[0])
```

```python
decomp_df.head()
```

```python
decomp_df.index=decomp_df['Month']
```

```python
decomp_df.sort_index(inplace= True)
```

```python
decomp_df['MilesMM'].fillna(value = decomp_df['MilesMM'].mean(), inplace=True)
```

```python
result = seasonal_decompose(decomp_df['MilesMM'], model='additive')
```

```python
result.plot()
```

```python
result2 = seasonal_decompose(decomp_df['MilesMM'], model='multiplicative')
```

```python
result2.plot()
```
